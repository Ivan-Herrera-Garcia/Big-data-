```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#Cada import importa una libreria en el orden que se escribe en las "variables" que estana al derecha
ruta = "D:\\procesamiento\\"    #Ruta donde estan guardados los archivos del dataset
```

# CIRCUITO A_CIRCUITO_A1 Y B_CIRCUITO_A1


```python
Archivo_A = pd.read_csv(ruta+"A_Circuito_A1.AV") #Circuito A
Archivo_B = pd.read_csv(ruta+"B_Circuito_A1.AV") #Circuito B
#Se crea A con los datos del circuito A
A = pd.DataFrame(Archivo_A)
A = A.dropna(how='all') #Se eliminan los NaN
A['Values'] = abs(A['Values']) #Se transforman los negativos a positivos
#Se crea B con los datos del circuito B
B = pd.DataFrame(Archivo_B)
B = B.dropna(how='all') #Se eliminan los NaN
B['Values'] = abs(B['Values']) #Se transforman los negativos a positivos
Resultado = A.merge ( B, on = "Datetime", suffixes = ( "-B-Circuito_A1", "-A-Circuito_B1" ) ) #Se juntan los dos DataFrame en Resultado
Resultado = Resultado.groupby ( "Datetime" ).mean()
#groupby Una operación groupby implica alguna combinación de dividir el objeto, aplicar una función y combinar los resultados.
# mean() Se crea una columna en donde estara el promedio
Resultado['PROMEDIO'] = Resultado.sum(axis=1).div(2) #Se crea una columna en donde estara el promedio
Resultado['SUMA'] = Resultado[["Values-B-Circuito_A1","Values-A-Circuito_B1"]].sum(axis=1) #Se crea una columna en donde estara la suma
Resultado['Desv_est'] = Resultado[["Values-B-Circuito_A1","Values-A-Circuito_B1"]].std(axis=1)
Resultado['Varianza'] = Resultado[["Values-B-Circuito_A1","Values-A-Circuito_B1"]].var(axis=1)
Resultado = Resultado.reset_index().copy()
#reset_index() Restablecer el índice de un DataFrame usando el método
#Retorna una copia superficial de x.
Resultado
```

# CIRCUITO A_CIRCUITO_04 Y B_CIRCUITO_04

```python
Archivo_A = pd.read_csv(ruta+"A_Circuito_04.AV") #Circuito A
Archivo_B = pd.read_csv(ruta+"B_Circuito_04.AV") #Circuito B
#Se crea A con los datos del circuito A
A = pd.DataFrame(Archivo_A)
A = A.dropna(how='all') #Se eliminan los NaN
A['Values'] = abs(A['Values']) #Se transforman los negativos a positivos
#Se crea B con los datos del circuito B
B = pd.DataFrame(Archivo_B)
B = B.dropna(how='all') #Se eliminan los NaN
B['Values'] = abs(B['Values']) #Se transforman los negativos a positivos
Resultado = A.merge ( B, on = "Datetime", suffixes = ( "-B-Circuito_04", "-A-Circuito_04" ) ) #Se juntan los dos DataFrame en Resultado
Resultado = Resultado.groupby ( "Datetime" ).mean()
#groupby Una operación groupby implica alguna combinación de dividir el objeto, aplicar una función y combinar los resultados.
# mean() Se crea una columna en donde estara el promedio
Resultado['PROMEDIO'] = Resultado.sum(axis=1).div(2) 
Resultado['SUMA'] = Resultado[["Values-B-Circuito_04","Values-A-Circuito_04"]].sum(axis=1) #Se crea una columna en donde estara la suma
Resultado['Desv_est'] = Resultado[["Values-B-Circuito_04","Values-A-Circuito_04"]].std(axis=1)
Resultado['Varianza'] = Resultado[["Values-B-Circuito_04","Values-A-Circuito_04"]].var(axis=1)
Resultado = Resultado.reset_index().copy()
#reset_index() Restablecer el índice de un DataFrame usando el método
#Retorna una copia superficial de x.
Resultado
```

```python
#Grafica para demostrar los valores de medicion del circuito A respecto al tiempo
Resultado [ [ "Datetime", "Values-A-Circuito_04" ] ].plot ( "Datetime", figsize = ( 15, 8 ), kind= 'kde') #Se eligio este 
#tipo de grafica para ver el manejo de matplotlib
graficaA = plt.subplot ( ) #graficaA permite la graficacion de tiempo-valorA
graficaA.set_ylabel ( "Estampa de tiempo" )  #Etiqueta en y
graficaA.set_xlabel ( "Valores de A" )       #Etiqueta en x
plt.title ( "Circuito A [Valores A x Estampa de tiempo]" ) #Nombre de la grafica
plt.show ( ) #Grafica lo que existe en plt
```

```python
#Grafica para demostrar los valores de medicion del circuito A respecto al tiempo
Resultado [ [ "Datetime", "Values-B-Circuito_04" ] ].plot ( "Datetime", figsize = ( 15, 8 ), kind= 'line') #Se eligio este 
#tipo de grafica para ver el manejo de matplotlib
graficaA = plt.subplot ( ) #graficaA permite la graficacion de tiempo-valorB
graficaA.set_ylabel ( "Estampa de tiempo" )  #Etiqueta en y
graficaA.set_xlabel ( "Valores de B" )       #Etiqueta en x
plt.title ( "Circuito B [Valores B x Estampa de tiempo]" ) #Nombre de la grafica
plt.show ( ) #Grafica lo que existe en plt
```

```python
#Grafica para la varianza
Resultado [ [ "Datetime", "Varianza" ] ].plot ( "Datetime", figsize = ( 15, 8 ), kind= 'area') #Se eligio este 
#tipo de grafica para ver el manejo de matplotlib
graficaA = plt.subplot ( ) #graficaA permite la graficacion de tiempo-varianza
graficaA.set_ylabel ( "Estampa de tiempo" )  #Etiqueta en y
graficaA.set_xlabel ( "Varianza" )           #Etiqueta en x
plt.title ( "Varianza" )                     #Nombre de la grafica
plt.show ( )                                 #Grafica lo que existe en plt
```

```python
#Covarianza
print(Resultado[["Values-A-Circuito_04","Values-B-Circuito_04"]].cov()) #Se imprime la tabla de covarianza
#Se hablo con la maestra debido a que no era facil graficar la covarianza por los valores que regresa la funcion :)
#En cambio se puso la graficacion de tiempo-valorA de cada circuito, saludos! 
```

```python
#Grafica para la desviacion estandar
Resultado [ [ "Datetime", "Desv_est" ] ].plot ( "Datetime", figsize = ( 15, 8 ), kind= 'density') #Se eligio este 
#tipo de grafica para ver el manejo de matplotlib
graficaA = plt.subplot ( ) #graficaA permite la graficacion de tiempo-desviacion-estandar
graficaA.set_ylabel ( "Estampa de tiempo" )  #Etiqueta en y
graficaA.set_xlabel ("Desviacion estandar")  #Etiqueta en x
plt.title ( "Desviacion estandar")           #Nombre de la grafica
plt.show ( )                                 #Grafica lo que existe en plt
```
